```python
import pandas as pd

from config.experiment_config import ExperimentConfig
```

```python
# N_ROWS = 1_000
N_ROWS = None

if N_ROWS:
    stocks_r = pd.read_csv(
        ExperimentConfig.PATH_MARKET_DATA / ExperimentConfig.RETURNS_FILENAME,
        nrows=N_ROWS,
    )
else:
    stocks_r = pd.read_csv(
        ExperimentConfig.PATH_MARKET_DATA / ExperimentConfig.RETURNS_FILENAME
    )
stocks_r = stocks_r.rename(columns={"caldt": "date"})
stocks_r["date"] = pd.to_datetime(stocks_r["date"])
stocks_r = stocks_r.set_index("date")
# presence_matrix = pd.read_csv(ExperimentConfig.PATH_MARKET_DATA / ExperimentConfig.PRESENCE_MATRIX_FILENAME, nrows=1_000)
# stocks_r.shape, presence_matrix.shape
stocks_r = stocks_r.add(1).resample("W").prod().add(-1)
stocks_r.head()
```

```python
from scipy.stats import skew, kurtosis

data_df = pd.read_csv(
    ExperimentConfig.PATH_OUTPUT / ExperimentConfig.INITIAL_FEATURES_FILENAME
)
data_df["date"] = pd.to_datetime(data_df["date"])
data_df = data_df.set_index("date")
data_df["spx_realized_var"] = data_df["spx"].resample("W-MON").var()
data_df["spx_realized_skew"] = data_df["spx"].resample("W-MON").aggregate(skew)
data_df["spx_realized_kurtosis"] = data_df["spx"].resample("W-MON").aggregate(kurtosis)
data_df = data_df.ffill()
data_df = data_df.resample("W-MON").last()
data_df = data_df[
    ["spx", "spx_realized_var", "spx_realized_skew", "spx_realized_kurtosis", ""]
]
data_df
```

```python
data_df.isna().any().any()
```

```python
stocks = list(stocks_r.columns)

full_df = pd.merge_asof(
    stocks_r.loc[data_df.index.min() :],
    data_df,
    left_index=True,
    right_index=True,
    direction="backward",
)
full_df.iloc[:, -1].head()
```

```python
with open(
    ExperimentConfig.PATH_OUTPUT / ExperimentConfig.STOCKS_LIST_FILENAME, mode="w"
) as f:
    f.write(",".join(stocks))
stocks
```
