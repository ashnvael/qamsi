```python
from pathlib import Path

import pandas as pd

from qamsi.utils.data import read_csv
from run import Dataset

TOP_N = 30
DATASET = Dataset.TOPN_US.value(topn=TOP_N)

recent = read_csv(".", f"QIS_targets_{TOP_N}.csv", "start_date")
```

```python
recent
```

```python
recent = recent.rename(columns={"start_date": "date", "shrinkage": "qis_shrinkage", "vol": "qis_vol"})
```

```python
data_df = pd.read_csv(Path("../../data/output") / DATASET.DF_FILENAME)
data_df["date"] = pd.to_datetime(data_df["date"])
data_df = data_df.set_index("date")
```

```python
data_df.shape
```

```python
data_df = data_df.merge(recent[["qis_shrinkage", "qis_vol"]], how="left", left_index=True, right_index=True)
data_df
```

```python
data_df.to_csv(Path("../../data/output") / DATASET.DF_FILENAME)
```
